```python
import numpy as np
import pandas as pd 
import heapq
import matplotlib.pyplot as plt
```

```python
#Import Trainning set and Test Set.
df = pd.read_csv('spam_train.csv')
df.head()
```

```python
#training set and test set 
X_train = df.drop('class',axis = 1 ).values
y_train = df['class']
target = df['class']
```

```python
df2 = pd.read_csv('spam_test.csv')
df2.head()
X_test = df2.drop('Label',axis = 1 ).drop(' ID',axis = 1).values
y_test = df2['Label']
Label = list(df2['Label'])
k = [ 1, 5, 11, 21, 41, 61, 81, 101, 201, 401]
```

```python
def z_score(X):
    x = X.T
    for i in range(len(x)):
        x[i] = (x[i]-x[i].mean())/x[i].std()
    return x.T
```

```python
def score(ypred,testing):
    return (list(np.array(ypred) - np.array(testing)).count(0))/len(ypred)  
```

```python
def get_predict(final,cls,k_num):
    list_ = []
    for i in final:
        s = 0 
        for n in i :
            s += cls[n]
        if s*2 > k_num:
            yred = int(1)
        else:
            yred = int(0)
        list_.append(yred)
    return list_
    
```

```python
def get_knn_predict(Xr,yr,Xt,yt,k_n,standard):
    Class = list(yr)
    Label = list(yt)
    yr = yr.values.reshape(-1,1)
    yt = yt.values.reshape(-1,1)
    if standard == True:
        xr = z_score(Xr)
        xt = z_score(Xt)
    else:
        xr = Xr.copy()
        xt = Xt.copy()
    Matrixdistance = []
    for j in range(len(xt)):
        listdistance = []
        outcome = []
        for u in range(len(xr)):
            distance = (((X_test[j]-X_train[u])**2).sum())**0.5
            listdistance.append(distance)
        Matrixdistance.append(listdistance)
    Database = pd.DataFrame(Matrixdistance) 
    list_acc = []
    for k in k_n:
        final_answer = []
        outcome = []
        y_pred = []
        for z in range(Database.shape[0]):
            list_distance = Database.iloc[z]
            target_distance = heapq.nsmallest(k,list_distance)
            min_num_index_list = list(map(list(list_distance).index, target_distance))
            final_answer.append(min_num_index_list)
        k_list = get_predict(final_answer,Class,k)
        acc = score(k_list,Label)
        list_acc.append(acc)
    return list_acc
```

```python
#unstandardizing
get_knn_predict(X_train,y_train,X_test,y_test,k_n = k,standard=False)
```

```python
#Standardizing
get_knn_predict(X_train,y_train,X_test,y_test,k_n = k,standard=True)
```
